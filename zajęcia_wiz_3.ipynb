```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dane00=pd.read_csv('citiesPL.csv')
```

```python
dane00['City']
```

```python
minW=13.91
maxE=24.73
minS=48.925
maxN=55.095
dane=dane00
plt.figure(figsize=(16,9))
nazwa=dane['City']
rozmiar=(dane['Population2021']*0.00025)
plt.scatter(dane['Longitude'],dane['Latitude'], s=(dane['Population2021']*0.00025), c='red',alpha=0.5)
plt.title('Mapa miast Polski')
plt.xlabel('Długość geograficzna')
plt.ylabel('Szerokość geograficzna')

for it, txt in enumerate(nazwa):
  plt.annotate(txt,(dane['Longitude'][it], dane ['Latitude'][it]+0.1), fontsize=0.065*rozmiar[it])

plt.annotate('Malwina',(18,52))
obraz=plt.imread('PL_tlo_800.png')
plt.imshow(obraz, extent=[minW,maxE,minS,maxN], alpha=0.75)
plt.xlim(minW,maxE)
plt.ylim(minS,maxN)
plt.axis('off')
srodekMapy=[minW+(maxE-minW)/2,minS+(maxN-minS)/2]
plt.show()
```

```python
import folium
data=dane00
map_poland= folium.Map(location=[srodekMapy[1],srodekMapy[0]],zoom_start=5.5)
for _, row in data.iterrows():
  folium.CircleMarker(
      location=[row['Latitude'], row['Longitude']],
      radius=10,
      color='red',
      popup=f"{row['City']}, {row['Population2021']}",
      fill=True,
      fill_opacity=0.6
  ).add_to(map_poland)


map_poland
```

```python
def lepsza_linia(dane00):
  startowa=10000000000
  lepszeLAT=0
  for it1 in np.linspace(minS, maxN, 1000):
    dane02=dane00[dane00['Latitude']>it1]
    wartosc1=sum(dane02["Population2021"])
    dane03=dane00[dane00['Latitude']<it1]
    wartosc2=sum(dane03["Population2021"])
    roznica=abs(wartosc1-wartosc2)
    if roznica<startowa:
      startowa = roznica
      lepszeLAT= it1
  return [lepszeLAT,roznica]
```

```python
lepsza_linia(dane00)
```
