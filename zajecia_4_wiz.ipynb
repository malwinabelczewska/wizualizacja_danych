```python
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt

f=open("/pan_wolodyjowski.txt",'r', encoding='utf-8')
tekst=f.read().split()
f.close()

from collections import Counter
def count_words(word_list):
  return Counter(word_list)

import string
lista_slow=[]
for word in tekst:
  if word !='—':
    for punctuation in string.punctuation:
      word=word.replace(punctuation,'')
    word=word.lower()
    if len(word)>=5:
      lista_slow.append(word)
  else:
    continue

word_counts=count_words(lista_slow)
print(word_counts.most_common())
x=word_counts.most_common(20)

slowa=[item[0] for item in x]
liczby=[item[1] for item in x]

plt.figure(figsize=(10,6))
cmap=plt.cm.get_cmap('plasma')
norm=mcolors.Normalize(vmin=min(liczby), vmax=max(liczby))
colors=[cmap(norm(value)) for value in liczby]
plt.bar(slowa, liczby, color=colors, width=0.7)
plt.xticks(rotation=90)
for i, v in enumerate(liczby):
  plt.text(i, v, str(v), ha='center', va='bottom', alpha=0.8)
plt.title("20 najczęściej występujących słów")
xlabell=plt.xlabel("Słowa", size=15)
ylabell=plt.ylabel("Liczba wystąpień", size=15)
plt.show
```

```python
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt

f=open("/pan_tadeusz.txt",'r', encoding='utf-8')
tekst=f.read().split()
f.close()

from collections import Counter
def count_words(word_list):
  return Counter(word_list)

import string
lista_slow=[]
for word in tekst:
  if word !='—':
    for punctuation in string.punctuation:
      word=word.replace(punctuation,'')
    word=word.lower()
    if len(word)>5:
      lista_slow.append(word)
  else:
    continue

word_counts=count_words(lista_slow)
print(word_counts.most_common())
x=word_counts.most_common(20)

slowa=[item[0] for item in x]
liczby=[item[1] for item in x]
slowa=slowa[::-1]
liczby=liczby[::-1]
plt.figure(figsize=(10,6))
cmap=plt.cm.get_cmap('cividis')
norm=mcolors.Normalize(vmin=min(liczby), vmax=max(liczby))
colors=[cmap(norm(value)) for value in liczby]
plt.barh(slowa, liczby, color=colors)
plt.xticks(rotation=90)
for i, v in enumerate(liczby):
  plt.text(v, i, str(v), ha='left', va='center', alpha=0.8)
plt.title("20 najczęściej występujących słów")
ylabell=plt.xlabel("Słowa", size=15)
xlabell=plt.ylabel("Liczba wystąpień", size=15)
plt.show
```
