```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter([2,4,6],[5,15,19],
          color='red',
          marker='*')
ax.set_xlim(1.5,6.5)
plt.show()
```

```python
fig, ax=plt.subplots()
x=np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')
ax.annotate('lokalne maksimum', xy=(2*np.pi,1), xytext=(10,4), 
           arrowprops=dict(facecolor='red', shrink=0.05))
ax.annotate('lokalne minimum', xy=(5*np.pi,-1), xytext=(2,-6),
           arrowprops=dict(arrowstyle='->',connectionstyle="angle3,angleA=0,angleB=-90"))
```

```python
x=np.linspace(-3,3,1000)
y=x**3-x**2-3*x+5
fig=plt.figure()
ax=fig.add_subplot()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.grid()
ax.set_ylim(-10,8)
ax.set_xlim(-3,3)
ax.plot(x,y,color='green')
plt.title(r"Wykres funkcji $f(x)=x**3-x**2-3*x+5$")
plt.xlabel("$x$",fontsize=10)
plt.ylabel("$f(x)$",fontsize=10)
```
