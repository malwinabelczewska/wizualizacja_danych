```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 1000)
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlim([0, 2])
plt.ylim([0, 10])
plt.savefig('przyklad02.pdf')
```
